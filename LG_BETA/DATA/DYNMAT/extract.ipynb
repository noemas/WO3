```python
import numpy

#name of input file containing dynamical matrix and eigenvectors
fildyneig_name = 'P4ncc_0.dyn'

#name of output file containing dynamical matrix
fildyn_name = 'dynmat_0'

#name of output file containing eigenvectors
fileig_name = 'eigenvector_0'

#number of line before dynamical matrix starts (line of last atomic coordinates)
dyn_start = 25

#index of dynmat to be extracted
idyn = 1

#number of dynmats
ndyn = 1

#________________________________________________________________________#


fildyneig = open(fildyneig_name, 'r')
fildyn = open(fildyn_name, 'w')
fileig = open(fileig_name, 'w')

nat = 0

#extract dynamical matrix
#skip first lines and get number of atoms
for i in range(0, dyn_start):
    line = fildyneig.readline()
    if i == 2:
        nat = int(line.split()[1])

#iterate through all dynmats
for i in range(0, ndyn):

    for j in range(0, 5):
        fildyneig.readline()
    
    for j in range(0, nat):
        for k in range(0, nat):
            fildyneig.readline()
            for l in range(0, 3):
                line = fildyneig.readline()
                if i == idyn - 1:
                    fildyn.write(line)
                    
#extract eigenvectors

for i in range(0, 6):
    fildyneig.readline()
    
for i in range(0, 3*nat):
    fildyneig.readline()
    for j in range(0, nat):
        line = fildyneig.readline()
        line = line.replace('(', '')
        line = line.replace(')', '')
        fileig.write(line)

fildyneig.close()
fildyn.close()
fileig.close()
        
```
