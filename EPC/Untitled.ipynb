```python
#Raub (1964)
#==========#
#first observation of superconductivity in NaWO3
#tetragonal I structure with x = [0.28, 0.35]
#no superconductivity outside this x-range
#Tc generally below 1K (one is reported at 0.57K but without x)
#https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.13.746

#Sweedler (1965)
#==========#
#superconductivity for MWO3, M = Na, K, Rb, Cs
#only superconductive in hexagonal and tetragonal I phases
#no superconductivity for NaWO3 in cubic and tetragonal II phase
#Reported Tc's:
#    - tetragonal NaWO3: x = 0.2, Tc=0.55 K 
#    - hexagonal KWO3: x = [0.27, 0.31], Tc = 0.5
#    - tetragonal KWO3: x = [0.4, 0.57], Tc = 1.5
#    - hexagonal RbWO3: x = [0.27, 0.29], Tc = 1.98
#    - hexagonal CsWO3: x = 0.32, Tc = 1.12
#No mention of Tc(x) behaviour so there might be a peak in the Tc(x) for KWO3
#https://www.sciencedirect.com/science/article/pii/0031916365912928?via%3Dihub%22

#Sweedler (1965)
#==========#
#superconductivity in tetragonal I BaWO3 (Ba donates 2 electrons)
#x = 0.13, Tc = 1.9
#synthesis of cubic NaWO3 with x=0.3 and x=0.4 with no superconductivity
#https://www.sciencedirect.com/science/article/pii/0031916365907171?via%3Dihub

#Hubble (1971)
#==========#
#increasing Tc with decreasing x observed in hexagonal K, Rb, and CsWO3
#However no sharp additional dip in Tc observed as in following years
#https://www.sciencedirect.com/science/article/pii/S0022369771802287

#Shanks (1974)
#==========#
#tetragonal I structure in NaWO3 for x = [0.2, 0.5]
#below x = 0.2 there is a transition to tetragonal II
#superconductivity in tetragonal I structure where Tc(x) is proportional to exp(-1/x)
#possible explanation in phonon mode softening with decreasing x
#Cooper pair energy (and to a first approximation) Tc is proportional to exp(-1/(V_ep*n_F))
#Thus if n_F ~ n, then V_ep ~ 1/x^2
#documentation of the transition mode (O-octahedra rotation, tetragonal Z3+, cubic R5+ mode)
#(not sure about the mode description - looks strange)
#https://www.sciencedirect.com/science/article/pii/0038109874902543

#Ngai (1976)
#==========#
#same description of the tetragonal TI-TII transition mode (as Shanks (1974))
#linear dependency of squared soft mode frequency on x
#inclusion of effect of soft phonon tunneling at transition x on soft mode frequency
#free energy difference between TI and TII assumed to be linear in x
#assumption of tunneling amplitude as an exponential of -x
#electron DOS treated as linear in x over wide range
#electron-phonon matrix elements treated as independent of x
#assumption that neither relevant electron nor phonon bands change much with x
#linear dependence of mu* on x
#good agreement with this model and Allen-Dynes Tc with Shanks (1974) data
#https://journals.aps.org/prb/abstract/10.1103/PhysRevB.13.1032

#Ngai (1978)
#==========#
#TI-TII transition soft mode apparently not observed in experiments (Kamitakahara?)
#new model of local structural excitations i.e. areas in TI where x < x_c
#causes tunneling between the structures (no phonons involved!)
#https://iopscience.iop.org/article/10.1088/0305-4608/8/1/018/meta


#Skokan (1979)
#==========#
#increase of Tc with decreasing x in hexagonal CsWO3
#however there is an additional dip in Tc (but no structural change observed by x-ray)
#hexagonal phase not stable anymore at lowest x where Tc is highest
#https://journals.aps.org/prb/abstract/10.1103/PhysRevB.20.3670

#Cadwell (1980)
#==========#
#increase of Tc with decreasing x in KWO3
#additional dip in Tc
#https://journals.aps.org/prb/pdf/10.1103/PhysRevB.23.2219

#Sato (1983)
#==========#
#find T-dependent softening phonon mode with decreasing at Gamma 
#but no soft mode in inelastic neutron scattering as presumed in model from Shanks (1974)
#use LSE model from Ngai
#good recollection of the papers before
#https://iopscience.iop.org/article/10.1088/0022-3719/16/27/007

#Garcia-Ruiz (1992)
#==========#
#different x-ranges for tetragonal I and cubic phases (cubic starts at x=0.28)
#two phases are coexisting
#https://www.sciencedirect.com/science/article/pii/092145349290575W?via%3Dihub

#Garif'yanov (1996)
#==========#
#synthesis of NaWO3 cubic crystal with conductivity in x = [0.16, 0.4]
#This x-range is different to Sweedler (1965)
#single crystals "collaps" below x = 0.16
#shows same inverse behaviour of Tc(x) on x: approx. -3.66x+2.24 in x = [0.15, 0,3]
#possible explanation in phonon mode softening at transition
#https://link.springer.com/article/10.1007/BF02583735

#Brusetti (2002)
#==========#
#increase of Tc with decreasing x in hexagonal RbWO3
#however there is an additional dip in Tc due to ordering of the Rb vacancies (no structual change observed)
#some transition implied at lowest x where Tc is highest
#https://journals.aps.org/prb/abstract/10.1103/PhysRevB.65.144528

#Bocarsly (2013)
#==========#
#increase of Tc with decreasing x in hexagonal InWO3
#highest Tc at x where new phase occurs
#similar behaviour to hexagonal CsWO3 and RbWO3 with additional dip in Tc
#https://iopscience.iop.org/article/10.1209/0295-5075/103/17001/pdf

#Haldolaarachchige (2014)
#==========#
#increase of Tc with decreasing x in tetragonal KWO3
#highest at x=0.35 which agrees with Sweedler (1965)
#https://journals.aps.org/prb/pdf/10.1103/PhysRevB.89.104520
```
